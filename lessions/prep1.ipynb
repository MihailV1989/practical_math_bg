Като задействаш блока с код по-горе, под него ще се появи като резултат въведения текст. В блока също така има коментар започващ с "#", който пояснява кода под него.

### 2 Прости математически операции
Простите математически операции като събиране, изваждане, умножение и деление се въвеждат супер лесно използвайки символите "+", "-", "*" и "/", примерно:

```python
a1 = 3 + 1
print('a1 = ', a1)

a2 = 2 * 2
print('a2 = ', a2)

a3 = 4 / 2
print('a3 = ', a3)

a = a1 - a2 + a3
print('a = ', a)
```

На това място ще е добре да споменя, че тук математическите операции се извършват една по една. Примерно когато първо разделим числото $12$ на $2$ и след това го умножим по $3$:

```python
b = 12 / 2 * 3
print('b = ', b)
```

 получаваме $18$, а не на $2$. Тоест умножението по $3$ се извършва чак след като делението на $2$ е извършено. Ако всъщност искаме да умножим делитела $2$ по $3$, трябва да използваме скоби ето така:

```python
b = 12 / (2 * 3)
print('b = ', b)
```

Изключение от това правило е повдигането на степен, което се извършва с две звездички "**":

```python
b = 12 / 2 ** 3
print('b = ', b)
```

Вместо да смятаме директно с числа, можем да смятаме и със съществуващи променливи. На пример в следната формула:

\begin{align}
    Z = f\cdot\frac{L}{L'} \tag{1}
\end{align}

ако за $f$, $L$ и $L'$ вече да са декларирани променливи със съответните стойности:

```python
f = 3000
L = 0.1
L_proj = 100
```

можем да пресметнем $Z$ използвайки променливите:

```python
Z = f * L / L_proj
print('Z = ', Z)
```

Освен това при програмирането винаги се използва десетична точка, вместо десетична запетая:

```python
c = 1.5 + 2
print('c = ', c)
```

тъй като запетаята се използва за изброяване, примерно на елементите на един списък:

```python
spisuk_pokupki = ['хляб', 'мляко', 'яйца']
print('Списък за пазаруване: ', spisuk_pokupki)
```

за които ще разбереш повече в следващата точка.

### 3 Работа със списъци (lists)
По-горе видя как се създава един списък (на английски list). Списъците се използват за да се запаметят повече от една стойност в една променлива, като стойностите могат да бъдат от различен тип, примерно както числа, така и текст. Така вместо да създаваме четири променливи за запаметяване на координатите на двете крайни точки на една линия $(x_1, y_1)$ и $(x_2, y_2)$:

```python
x1 = 2
y1 = 1
x2 = 6
y2 = 7
```

можем да запаметим всички в един единствен списък:

```python
coordinates1 = [x1, y1, x2, y2]
print('Координати на точките на линията: ', coordinates1)
# или ако директно въведем стойностите:
coordinates2 = [2, 1, 6, 7]
print('Координати на точките на линията: ', coordinates2)
```

след което можем да изкараме отделните стойности от списъка като посочим индекса на желания елемент:

```python
print('x1 = ', coordinates2[0])
print('y1 = ', coordinates2[1])
print('x2 = ', coordinates2[2])
print('y2 = ', coordinates2[3])
```

Тук трябва да обърнеш внимание, че индексите (тоест номерацията) на елементите на един списък започват от нула, а не от едно.  

Тъй като искаме да запаметим $(x, y)$ координатите на две точки, би било практично ако групираме координатите според това дали са на първата или втората точка на линията. За целта можем да създадем два отделни списъка за двете точки в основния списък ето така:

```python
coordinates = [[x1, y1], [x2, y2]]
print('Координати на точките на линията: ', coordinates)
```

Сега ако изкараме първия елемент на списъка, ще получим координатите на първата точка на линията $[x_1, y_1]$:

```python
print('Координати на първата точка на линията: ', coordinates[0])
```

Ако искаме да прочетем $x$ и $y$ компонентите на първата точка по отделно, то трябва от първия елемент `[0]` на променливата `coordinates` да изкараме първия `[0]` и последния `[1]` елемент ето така:

```python
print('x компонентата на първата точка на линията: ', coordinates[0][0])
print('y компонентата на първата точка на линията: ', coordinates[0][1])
```

Тоест можем да работим със списъци в списъци, които са на няколко нива, като чрез многократно използването на квадратните скоби `[]` имаме достъп до всеки елемент на всеки от списъците.

За по-прегледно можем да използваме втори вид списъци, които се задават с кръгли скоби `()` и се различават от първия вид списъци с квадратни скоби `[]` по това, че веднъж създадени, те не могат да се променят. Така можем да използваме квадратни скоби `[]` за списъка с точките на линията и кръгли скоби `()` за координатите на съответната точка:

```python
coordinates = [(x1, y1), (x2, y2)]
print('Координати на точките на линията: ', coordinates)
```

### 4 Зареждане на библиотеки

За първата задача ни трябват библиотеки за обработка и показване на изображения, за интерактивно взаимодействие със скриптовете, за по-специални математически операции и няколко специално за задачата направени модула. Зареждаме ги с функцията `import` последвана от името на пакета/модула. Нужните библиотеки вече се въведени в блока по-долу и ти трябва само да го задействаш:

```python
# активиране на интерактивния графичен плот
%matplotlib widget
# зареждане на библиотеката за графичен плот
import matplotlib.pyplot as plt
# зареждане на библиотеката за интерактивни контроли
import ipywidgets as widgets
# зареждане на библиотеката OpenCV за компютърно зрение за обработка на изображение
import cv2
# зареждане на библиотеката за математически изчисления
import math
# зареждане функцията ruler_plt() от модула за измерване на дължини върху плот
from custom_scripts.measure_plt import ruler_plt
# зареждане на функцията read_img() от модула за прочитане на качени снимки
from custom_scripts.read_uploaded_img import read_img
# изкарване на съобщение като се извършат всички команди
print('>>>Готово!<<<')
```

С библиотеката за компютърно зрение OpenCV (импортирана по-горе като `cv2`) можем да зареждаме наличните снимки от виртуалната машина:

```python
# зареждане на снимката от файл 'lession1/calibration.jpg' в променливата image2
image2 = cv2.imread('lession1/calibration.jpg')
# преобразуване в стандартната за matplotlib цветова схема червено/зелено/синьо (RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
# изкарване на съобщение като се извършат всички команди
print('>>>Готово!<<<')
```

С функцията `ruler_plt()` можем да модифицираме плот, създаден с `plt`, така че да можем да измерваме дължини върху снимката с линия.

С него се работи по следния начин:
- поставяш началната и крайната точка на линията чрез натискане на десния бутон на мишката (съкратено като "RMB", от английски "right mouse button") върху снимката на желаното място
- след като вече е начертана линията, при повторно натискане на RMB върху изображението линията се изтрива и се създава нова начална точка за нова линия
- можеш да поставяш началната и крайната точка на линията на ново толкова пъти, колкото е нужно за да оцелиш точно желаната дължина.

Активирай блока по-долу за да изпробваш сам функцията:

```python
# създаване на нов плот
fig = plt.figure()
# създаване на подплот
ax = fig.add_subplot(111)
# показване на снимката върху плота
ax.imshow(image2)
# активиране на функцията за чертане на линия при кликване върху плота
coords = ruler_plt(fig, ax)
```

В блока по-горе, като резултат от функцията `ruler_plt()` получваме координатите на двете точки на начертаната от нас линия, които в случая запаметяваме в променливата `coords`. Като начертаеш линията, можеш с блока по-долу да покажеш координатите:

```python
print('Координати на линията: ', coords)
```
