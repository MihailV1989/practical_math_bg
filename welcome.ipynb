## Полезна математика
Тук ще намерите интерактивни задачи от сферата на компютърното зрение, за чиито решаване ще използваме наученото в часовете по математика в училище. Целта във всяка задача е да се реши някакъв практически проблем по много опростен начин, за да е достъпна и разбираема за учениците от горните класове. За да са интересни задачите за решаване, ще използваме езика за програмиране Python за демонстриране на резултатите. За това не се изискват никакви предварителни познания по програмиране, но пък при желание може да се възползвате от задачите за да трупате знания и по програмиране. Освен това в задачите се разясняват различни понятия от оптиката и фотографията, които са нужни за разбирането и решаването на задачите.
