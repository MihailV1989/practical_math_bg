## Полезна математика
Тук ще намериш интерактивни задачи от сферата на компютърното зрение, за чиито решаване ще използваме наученото в часовете по математика в училище. Целта във всяка задача е да се реши някакъв практически проблем по много опростен начин, за да е достъпна и разбираема за учениците от горните класове. За да са интересни задачите за решаване, ще използваме прекрасния език за програмиране Python за демонстриране на резултатите. За това не се изискват никакви предварителни познания по програмиране, но пък при желание може да се възползваш от задачите за да трупаш знания и по програмиране. Освен това в задачите се разясняват различни понятия от оптиката и фотографията, които са нужни за разбирането и решаването на задачите.

### Защо точно математика? Защо хората си причиняват главоболие с нея?
Математиката е изключително логическа и последователна наука, която ни позволява да описваме закономерностите от другите науки на лист хартия. С нейна помощ във физиката описваме света около нас. Преподавана по подходящ начин, математиката развива критическото мислене, въображението и способността ни да решаваме сложни проблеми, които не можем да решим само със сръчност и майсторлък. Решаването на такива проблеми ни позволява да конструираме мощни двигатели, бързи коли, още по-бързи самолети и космически кораби и дори да намалим негативното влияние на човечеството върху природата, въздуха, който дишаме, храната, която ядем и водата, която пием.

Обаче! Математиката често е доста сложна и абстрактна. Отвъд събирането на ябълки и разделенето на торта на равни части е много трудно да се обясни математиката по нагледен начин с примери от всекидневието ни. В училище някъде след 7 клас математиката мен ме загуби. Беше ми трудно да разбирам какво смятаме, а имах желание. Доста години по-късно, по стечение на обстоятелствата записах да следвам машиностроене. Налагаше ми се да се уча да използвам математиката за решаването на практични задачи от механиката, динамиката и други науки. Успях някак си да наваксам пропуснатите познания по математика, но хич не ми беше лесно. Малко по-малко почнах да се възхищавам от матемакита и да я обичам.
